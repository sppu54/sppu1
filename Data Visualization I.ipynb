```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\CHETAN\Desktop\Titanic-Dataset.csv')
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#calculating mean_values
avg_Age = df['Age'].astype(float).mean(axis = 0)
avg_Age = int(avg_Age)
avg_Age
```

```python
#replacing NaN values by mean_values
df['Age'].replace(np.nan,avg_Age,inplace = True)
```

```python
df.dropna(subset = ['Embarked'] , axis = 0, inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
df.head()
```

```python
mode_Cabin = df['Cabin'].mode()[0]
print(mode_Cabin)
```

```python
df['Cabin'] = df['Cabin'].fillna(mode_Cabin)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(data = df, x = 'Age', y = 'Fare')
```

```python
sns.scatterplot(data = df, x = 'Fare', y = 'Age', hue = 'Survived')
```

```python
sns.histplot(df['Age'], bins = 10)
```

```python
sns.displot(df['Age'])
```

```python
sns.barplot(x = 'Sex', y = 'Survived', data = df)
```

```python
sns.histplot(df['Sex'])
```

```python
sns.barplot(x = 'Sex', y = 'Survived', data = df, hue = 'Embarked')
```

```python
sns.pairplot(df)
```

```python
sns.histplot(df['fare'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Ticket Price')
plt.ylabel('Frequency')
plt.title('Distribution of Ticket Prices')
plt.show()
```
