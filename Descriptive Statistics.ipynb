```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
df = pd.DataFrame(iris['data'])
df.head()
```

```python
df[4] = iris['target']
df.head()
```

```python
# Adding column names
df.rename(columns = {0:'SepalLengthCm', 1:'SepalWidthCm', 2:'PetalLengthCm', 3:'PetalWidthCm', 4:'Species'}, inplace = True)
df.head()
```
